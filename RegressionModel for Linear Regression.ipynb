```python
#spending score as dependent var and annual income as independent variable
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("Dataset5.csv")
```

```python
#View the dataset
data
```

```python
data.shape
```

```python
data.plot(kind = 'scatter', x='Annual Income (k$)',y='Spending Score (1-100)')
plt.show()
```

```python
data.corr() #Correlation coefficients
```

```python
#change to dataframe variables
Annual_Income=pd.DataFrame(data['Annual Income (k$)'])
Spending_Score=pd.DataFrame(data['Spending Score (1-100)'])
```

```python
Annual_Income
```

```python
#Build linear regression model
lm = linear_model.LinearRegression()
model=lm.fit(Annual_Income,Spending_Score)
```

```python
model.coef_ #slope
```

```python
model.intercept_ #intercept
```

```python
model.score(Annual_Income,Spending_Score) #Evaluate the model
```

```python
data.plot(kind='scatter',x='Annual Income (k$)',y='Spending Score (1-100)')
#plotting the regression line
plt.plot(Annual_Income,model.predict(Annual_Income),color='red',linewidth=2)
#plotting the predict value
plt.show()
```
